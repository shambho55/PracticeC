## Bubble Graphs

```python
import matplotlib.pyplot as plt
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
plt.scatter(year,salary,s = 100,c = 'r')
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
c[0] = c[2]
c
```

```python
# c is color, s is size , alpha is for transparency. In color we can pass array of colors
plt.scatter(year,salary,s = population,c = c,alpha = 0.5)
plt.title('Bubble Chart')
plt.show()
```

```python
plt.scatter(year,salary,s = population,c = c,marker = '^',edgecolor = 'black',alpha = 0.5)
plt.title("Bubble Chart")
for i in range(len(year)):
    if i == len(year) - 1:
        plt.text(year[i] - 0.1,salary[i] - 0.9,population[i])
    else:
        plt.text(year[i] + 0.1,salary[i] + 0.1,population[i])
plt.show()
```

```python
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
c = np.arange(len(year))
plt.scatter(year,revenue,s = employees,c = c,edgecolor = 'black',alpha = 0.7)
plt.grid()
plt.title("Bubble Chart")
for i in range(len(year)):
    if i == len(year) - 1:
        plt.text(year[i] - 1,revenue[i] - 1,employees[i])
    else:
        plt.text(year[i] + 0.2,revenue[i] + 0.2,employees[i])
plt.show()
```

## Pie Graphs

```python
label = ['A','B','C','D']
sizes = [3,4,6,2]
colors = ['blue','purple','red','pink']
explode = [0.1,0.2,0,0]
plt.title('Split among classes')
plt.pie(sizes,colors = colors,explode = explode,labels = label, autopct = '%.2f%%',counterclock = False,startangle = 100)
plt.axis('equal')
plt.show()
```

```python
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
plt.pie(sold,labels = company,autopct = '%.1f',counterclock = False)
plt.axis('equal')
plt.show()
sam = sum(sold)
for i in range(len(company)):
    print(company[i],round((sold[i]/sam) * 100,1))
```

```python
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
sold = np.array(sold)
plt.pie(sold,labels = company,autopct = '%.1f',counterclock = False)
plt.axis('equal')
plt.show()
ans = np.true_divide(sold,sold.sum(axis = 0,keepdims = True))*100
for i in range(len(company)):
    print(company[i],format(ans[i],'.1f'))
```

```python
# In this case, width = (max - min) / bins
# Bins are in range from [a,b) where a is included but b is excluded 
# But in last case , both are included [a,b]
plt.hist(b,bins = 21,edgecolor = 'black')
plt.xticks(xt)
plt.show()
```
